```python
import requests

def get_crypto_prices(coin_ids, currency='usd'):
    url = 'https://api.coingecko.com/api/v3/simple/price'
    
    try:
        response = requests.get(url, params={'ids': ','.join(coin_ids), 'vs_currencies': currency})
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Ошибка при получении данных: {e}")
        return {}

coins = ['bitcoin', 'ethereum', 'solana', 'invalidcoin']
prices = get_crypto_prices(coins)

for coin in coins:
    price = prices.get(coin, {}).get('usd')
    print(f"{coin.capitalize()}: ${price:,.2f}" if price else f"{coin.capitalize()}: Цена недоступна")
```

```python
import pandas as pd
import string
from textblob import TextBlob
import seaborn as sns
import matplotlib.pyplot as plt

# Исходные данные
news_data = [
    {"title": "Bitcoin hits new all-time high!"},
    {"title": "Ethereum sees massive growth in Q1"},
    {"title": "Crypto market dips amid inflation fears"},
    {"title": "Solana outperforms major altcoins"},
    {"title": "Regulation fears impact crypto prices"}
]

df_news = pd.DataFrame(news_data)

df_news['clean_title'] = df_news['title'].str.lower().str.translate(str.maketrans('', '', string.punctuation))

df_news['sentiment'] = df_news['clean_title'].apply(lambda text: TextBlob(text).sentiment.polarity)

print(df_news[['title', 'sentiment']])

sns.histplot(df_news['sentiment'], bins=10, kde=True)
plt.title("Распределение тональности заголовков крипто-новостей")
plt.xlabel("Полярность тональности")
plt.ylabel("Частота")
plt.show()
```

```python
import pandas as pd
import requests

def fetch_historical_data(coin_id='bitcoin', vs_currency='usd', days=90):
    url = f'https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart'
    params = {
        'vs_currency': vs_currency,
        'days': str(days),  
        'interval': 'daily'
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  
        data = response.json()

        if 'prices' not in data:
            print("Ошибка: API не вернул данные о ценах")
            return pd.DataFrame()  

        df = pd.DataFrame(data['prices'], columns=['timestamp', 'price'])
        df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
        df.set_index('date', inplace=True)
        return df[['price']]

    except requests.exceptions.RequestException as e:
        print(f"Ошибка при запросе данных: {e}")
        return pd.DataFrame()  

df_prices = fetch_historical_data()
print(df_prices.head())
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import pandas as pd

series = df_prices['price']


model = ARIMA(series, order=(5, 1, 0))
model_fit = model.fit()


forecast = model_fit.forecast(steps=10)
print(forecast)


plt.figure(figsize=(10, 5))
plt.plot(series, label='Исторические данные')
plt.plot(pd.date_range(series.index[-1], periods=11, freq='D')[1:], forecast, label='Прогноз', color='red')
plt.title("Прогноз стоимости биткойна (ARIMA)")
plt.legend()
plt.show()
```
